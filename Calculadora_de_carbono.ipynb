```python
# ==============================================================================
# CALCULADORA DE PEGADA DE CARBONO
# ROTA: MANAUS -> MANACAPURU
# ==============================================================================

# DISTÂNCIAS (km)
DISTANCIA_TERRESTRE = 100
DISTANCIA_FLUVIAL = 102

# FATORES DE EMISSÃO (kg CO2/L)
FATOR_DIESEL = 2.68
FATOR_GASOLINA = 2.31
FATOR_ETANOL = 1.50

# CONSUMO MÉDIO (km/L)
CONSUMO_CARRO = 10
CONSUMO_ONIBUS = 4
CONSUMO_LANCHA = 2
CONSUMO_RECREIO = 0.8

# CAPACIDADE
CAP_CARRO = 5
CAP_ONIBUS = 45
CAP_LANCHA = 20
CAP_RECREIO = 200


# ==============================================================================
# FUNÇÕES
# ==============================================================================

def calcular_terrestre(tipo, passageiros, combustivel):

    if tipo == "carro":

        if passageiros > CAP_CARRO:
            return None

        consumo = CONSUMO_CARRO

        if combustivel == "gasolina":
            fator = FATOR_GASOLINA
        else:
            fator = FATOR_ETANOL

    elif tipo == "onibus":

        if passageiros > CAP_ONIBUS:
            return None

        consumo = CONSUMO_ONIBUS
        fator = FATOR_DIESEL

    else:
        return None

    litros = DISTANCIA_TERRESTRE / consumo
    co2_total = litros * fator

    return co2_total / passageiros


def calcular_fluvial(tipo, passageiros):

    if tipo == "lancha":

        if passageiros > CAP_LANCHA:
            return None

        consumo = CONSUMO_LANCHA
        fator = FATOR_GASOLINA

    elif tipo == "recreio":

        if passageiros > CAP_RECREIO:
            return None

        consumo = CONSUMO_RECREIO
        fator = FATOR_DIESEL

    else:
        return None

    litros = DISTANCIA_FLUVIAL / consumo
    co2_total = litros * fator

    return co2_total / passageiros


# ==============================================================================
# PROGRAMA PRINCIPAL
# ==============================================================================

print("\nCALCULADORA DE PEGADA DE CARBONO")
print("Rota: Manaus -> Manacapuru\n")

passageiros = int(input("Quantidade de passageiros: "))

resultados = {}

carro_gasolina = calcular_terrestre("carro", passageiros, "gasolina")
carro_etanol = calcular_terrestre("carro", passageiros, "etanol")
onibus = calcular_terrestre("onibus", passageiros, "diesel")

lancha = calcular_fluvial("lancha", passageiros)
recreio = calcular_fluvial("recreio", passageiros)

if carro_gasolina is not None:
    resultados["Carro (Gasolina)"] = carro_gasolina

if carro_etanol is not None:
    resultados["Carro (Etanol)"] = carro_etanol

if onibus is not None:
    resultados["Ônibus"] = onibus

if lancha is not None:
    resultados["Lancha"] = lancha

if recreio is not None:
    resultados["Barco de Recreio"] = recreio


if len(resultados) == 0:

    print("\nNenhum veículo suporta essa quantidade de passageiros.")

else:

    print("\nEmissão de CO₂ por passageiro:\n")

    for veiculo, valor in resultados.items():
        print(f"{veiculo}: {valor:.2f} kg CO₂")

    melhor = min(resultados, key=resultados.get)

    print(f"\nOpção mais sustentável: {melhor}")
```
